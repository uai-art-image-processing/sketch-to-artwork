## Data Exploration

```python
dataset = pd.read_csv("omniart_v3_datadump.csv", delimiter=",")
dataset.info()
```

```python
dataset.head()
```

### Unique Entries

```python
dataset.apply(pd.Series.nunique).to_frame()
```

### Artists

```python
# First 20
dataset['artist_full_name'].value_counts().to_frame().head(20)
```

### Sources

```python
dataset['collection_origins'].value_counts().to_frame()
```

```python
dataset['collection_origins'].value_counts().plot.barh(xlabel="collection", ylabel="size").set_title("Collection sizes")
```

### Artwork type

```python
# First 20
dataset['artwork_type'].value_counts().to_frame().head(20)
```

### School

```python
# First 20
dataset['school'].value_counts().to_frame().head(20)
```

```python
for i in dataset['school'].unique():
    print(i)
```

### general_type

```python
dataset['general_type'].value_counts().to_frame()
```

```python
dataset['general_type'].value_counts().drop(index=['unknown']).plot.barh(xlabel="Type", ylabel="size").set_title("General Type sizes")
```
